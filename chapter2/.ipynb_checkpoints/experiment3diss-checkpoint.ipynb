```python
# The true modelcen
percentage = .2
C1 = SFT('C1', ['normal' , 'infected'], ['L1', 'C2', 'R1'] ,
    {'L1': np.array([[1, 0], [1,percentage]]),
     'C2': np.array([[1, 0], [1,percentage]]),
     'R1': np.array([[1, 0], [1,percentage]])},
    ['clean', 'malicious'], 'external')

L1 = SFT('L1', ['normal' , 'infected'], [] ,
    {},
    ['clean', 'malicious'], 'external')

R1 = SFT('R1', ['normal' , 'infected'], [] ,
    {},
    ['clean', 'malicious'], 'external')

C2 = SFT('C2', ['normal' , 'infected'], ['L2', 'C3', 'R2'] ,
    {'L2': np.array([[1, 0], [1,percentage]]),
     'C3': np.array([[1, 0], [1,percentage]]),
     'R2': np.array([[1, 0], [1,percentage]])},
    ['clean', 'malicious'], 'external')

L2 = SFT('L2', ['normal' , 'infected'], [] ,
    {},
    ['clean', 'malicious'], 'external')

R2 = SFT('R2', ['normal' , 'infected'], [] ,
    {},
    ['clean', 'malicious'], 'external')

C3 = SFT('C3', ['normal' , 'infected'], ['L3', 'C4', 'R3'] ,
    {'L3': np.array([[1, 0], [1,percentage]]),
     'C4': np.array([[1, 0], [1,percentage]]),
     'R3': np.array([[1, 0], [1,percentage]])},
    ['clean', 'malicious'], 'external')

L3 = SFT('L3', ['normal' , 'infected'], [] ,
    {},
    ['clean', 'malicious'], 'external')

R3 = SFT('R3', ['normal' , 'infected'], [] ,
    {},
    ['clean', 'malicious'], 'external')

C4 = SFT('C4', ['normal' , 'infected'], ['L4', 'R4'] ,
    {'L4': np.array([[1, 0], [1,percentage]]),
     #'C5': np.array([[1, 0], [1,percentage]]),
     'R4': np.array([[1, 0], [1,percentage]])},
    ['clean', 'malicious'], 'external')

L4 = SFT('L4', ['normal' , 'infected'], [] ,
    {},
    ['clean', 'malicious'], 'external')

R4 = SFT('R4', ['normal' , 'infected'], [] ,
    {},
    ['clean', 'malicious'], 'external')

C5 = SFT('C5', ['normal' , 'infected'], [] ,
    {},
    ['clean', 'malicious'], 'external')

truenodes = [C1, R1, L1, C2, R2, L2, C3, R3, L3, C4,R4,L4,C5]
s0 = dict(zip(['C1', 'R1', 'L1', 'C2', 'R2', 'L2', 'C3', 'R3', 'L3', 'C4', 'R4', 'L4', 'C5'], 
              ['normal']*13))
s0['C1']= 'infected'
truenet =SFTNet(truenodes)
percore = 500
s_input = s0
def f1(x):
    return get_roc_coords(x, percore, percore, truenet, s0=s_input ,  \
                                  T=10, method='direct', directsamps=2500, printsteps=False)
from multiprocessing import Pool
p = Pool(8)
results1 = p.map(f1, list(np.random.randint(0, 20000, 8)))
pos_5, neg_5 = handle_parallel_res(results1)
our_roc_pts = np.asarray(plot_our_roc(pos_5, neg_5, .01))
anom_roc_pts = np.asarray(plot_anomaly_roc(pos_5, neg_5, .01))
plt.plot(our_roc_pts[:,0], our_roc_pts[:,1], label ='Likelihood Ratio Detector')
plt.plot(anom_roc_pts[:,0], anom_roc_pts[:,1], label = 'Baseline Anomaly Detector')
plt.xlim(0,1)
plt.ylabel('True Positive rate', fontsize=15)
plt.xlabel('False Positive rate', fontsize=15)
plt.legend(loc=0)
```
